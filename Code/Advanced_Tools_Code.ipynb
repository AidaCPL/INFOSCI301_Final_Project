```python
def align_climate_fever_schema(dataset):
    """Align Climate-FEVER dataset schema."""
    dataset = dataset.rename(columns={
        'claim_id': 'id',
        'claim': 'text',
        'claim_label': 'label',
        'evidences': 'evidence'
    })
    dataset['source'] = 'Climate-FEVER'
    return dataset

def align_global_misinformation_schema(dataset):
    """Extract relevant columns and align schema for Global Misinformation Dataset."""
    relevant_columns = ['Country', 'political', 'science', 'business', 'celebrity']
    dataset = dataset[relevant_columns]
    dataset = dataset.rename(columns={
        'Country': 'region',
        'political': 'political_engagement',
        'science': 'science_engagement',
        'business': 'business_engagement',
        'celebrity': 'celebrity_engagement'
    })
    return dataset

# Align schemas
climate_fever_aligned = align_climate_fever_schema(climate_fever)
global_misinformation_aligned = align_global_misinformation_schema(global_misinformation_dataset)

# Save aligned datasets
climate_fever_aligned.to_csv('climate_fever_aligned.csv', index=False)
global_misinformation_aligned.to_csv('global_misinformation_aligned.csv', index=False)
```

```python
# Combine all datasets into a unified format
combined_dataset = pd.concat([climate_fever_aligned,global_misinformation_aligned],
                              ignore_index=True)


# Save the integrated dataset
combined_dataset.to_csv('combined_dataset.csv', index=False)

# Add metadata from global misinformation dataset if required
# For example, merging by region
combined_with_behavior = pd.merge(combined_dataset, global_misinformation_aligned,
                                   left_on='region', right_on='region', how='left')

# Save the enriched dataset
combined_with_behavior.to_csv('combined_with_behavior.csv', index=False)
```

```python
import shap
import xgboost as xgb
import pandas as pd
import matplotlib.pyplot as plt

# Load the main dataset
main_data = global_misinformation_dataset

# Select relevant features
features = [
    "fb", "tw", "sn", "ig", "ti",
    "political", "imp_pol_aligned",
    "Edu", "risk", "trust",
    "urbanrural", "Minority", "age", "sex"
]

# Create a synthetic target for demonstration
# This can be replaced with an actual target column if available
main_data["misinformation_susceptibility"] = (
    main_data["political"] * 0.3 +
    main_data["imp_pol_aligned"] * 0.2 +
    main_data["trust"] * -0.2 +
    main_data["fb"] * 0.1 +
    main_data["Edu"] * -0.1 +
    main_data["risk"] * 0.2
)

# Define features (X) and target (y)
X = main_data[features]
y = main_data["misinformation_susceptibility"]
```

```python
# Check for NaN or infinite values in the dataset
print("Checking for NaN or Infinite Values...")
print("Features (X):")
print(X.isnull().sum())
print("Target (y):")
print(y.isnull().sum())

# Remove rows with NaN or infinite values
X = X.replace([float('inf'), -float('inf')], pd.NA).dropna()
y = y.replace([float('inf'), -float('inf')], pd.NA).dropna()

# Ensure X and y have the same indices after cleaning
common_indices = X.index.intersection(y.index)
X = X.loc[common_indices]
y = y.loc[common_indices]

# Train the XGBoost model again after cleaning
model = xgb.XGBRegressor(objective="reg:squarederror", n_estimators=100, max_depth=4, random_state=42)
model.fit(X, y)

# Perform SHAP analysis
explainer = shap.Explainer(model, X)
shap_values = explainer(X)

# Generate SHAP summary plot
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X, plot_type="bar")
plt.title("SHAP Feature Importance for Misinformation Susceptibility", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Extract raw SHAP values (matrix) and feature names from the SHAP explainer
shap_values_array = shap_values.values  # Raw SHAP values
feature_names = X.columns  # Feature names

# Generate a SHAP dependence plot for "risk"
shap.dependence_plot("risk", shap_values_array, X, feature_names=feature_names)
```
