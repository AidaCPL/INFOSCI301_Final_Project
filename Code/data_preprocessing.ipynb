## Network Preprocessing

```python
# Necessary libraries to import
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
import itertools
from itertools import combinations
from collections import defaultdict
from collections import Counter

# Download the required NLTK resources
nltk.download('punkt')          # Tokenizer for splitting text into words
nltk.download('stopwords')      # Stopwords for filtering out common words
nltk.download('wordnet')        # WordNet for lemmatization
nltk.download('punkt_tab')
nltk.download('omw-1.4')        # Required for WordNet Lemmatizer to function correctly
```

### NLP processes

```python
climate_fever_data =  climate_fever_data[climate_fever_data['evidence_label'].isin(['SUPPORTS', 'REFUTES'])]
```

```python
# Select the necessary columns
evidence_texts = climate_fever_data['evidence_text']
evidence_labels = climate_fever_data['evidence_label']

# Tokenization function
def tokenize(text):
    return text.split()

# Create co-occurrence pairs with labels
edges = []
for text, label in zip(evidence_texts, evidence_labels):
    tokens = tokenize(text)
    unique_pairs = combinations(set(tokens), 2)  # Generate unique word pairs
    for pair in unique_pairs:
        edges.append((*pair, label))

# Count edge weights
edge_weights = Counter(edges)

# Extract unique words as nodes
unique_words = set()
for text in evidence_texts:
    unique_words.update(tokenize(text))

# Create nodes DataFrame
nodes_data = pd.DataFrame({'Id': list(unique_words), 'Label': list(unique_words)})

# Create edges DataFrame
edges_data = []
for (source, target, label), weight in edge_weights.items():
    edges_data.append({'Source': source, 'Target': target, 'Weight': weight, 'Label': label})

edges_data = pd.DataFrame(edges_data)

# Export to CSV for Gephi
nodes_data.to_csv('nodes.csv', index=False)
edges_data.to_csv('edges.csv', index=False)
```

```python
# Define stop words
stop_words = set(stopwords.words('english'))

# Function to clean text and remove stop words
def remove_stop_words(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)

# Clean nodes in the edges file
edges_data['Source'] = edges_data['Source'].apply(remove_stop_words)
edges_data['Target'] = edges_data['Target'].apply(remove_stop_words)

# Filter out rows where the source or target becomes empty after cleaning
edges_data = edges_data[(edges_data['Source'] != '') & (edges_data['Target'] != '')]

# Create the graph
G = nx.DiGraph()  # Directed graph
for _, row in edges_data.iterrows():
    G.add_edge(row['Source'], row['Target'], weight=row['Weight'], label=row['Label'])

# 1. Remove low-weight edges (set a threshold)
weight_threshold = 5  # Adjust threshold as needed
G = nx.DiGraph([(u, v, d) for u, v, d in G.edges(data=True) if d['weight'] >= weight_threshold])

# 2. Remove low-degree nodes
degree_threshold = 2  # Minimum degree to keep
low_degree_nodes = [node for node, degree in G.degree() if degree < degree_threshold]
G.remove_nodes_from(low_degree_nodes)

# 3. Focus on the largest connected component (optional, for undirected graphs)
if not nx.is_connected(G.to_undirected()):
    largest_cc = max(nx.connected_components(G.to_undirected()), key=len)
    G = G.subgraph(largest_cc).copy()

# 4. Limit to top N nodes based on degree (optional)
N = 50  # Number of nodes to keep
top_nodes = sorted(G.degree, key=lambda x: x[1], reverse=True)[:N]
G = G.subgraph([node for node, _ in top_nodes]).copy()

# Save the simplified graph back to CSV for visualization
edges_simplified = pd.DataFrame({
    'Source': u,
    'Target': v,
    'Weight': d['weight'],
    'Label': d['label']
} for u, v, d in G.edges(data=True))

edges_simplified.to_csv('/content/edges_simplified_no_stopwords.csv', index=False) # Also available on GitHub

print("Simplified graph with stop words removed saved to 'edges_simplified_no_stopwords.csv'")
```
